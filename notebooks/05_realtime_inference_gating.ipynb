# 05_realtime_inference_gating.ipynb — Low-Latency Fast-Path Gating

In-match tactical decisions must be made under tight latency bounds. To prevent wasting inference cycles on cutscenes, pauses, or menu screens, we introduce a low-overhead gating check. This notebook explores fast heuristics (such as dominant HSV pitch color ratios or simple HUD pixel checks) to dynamically route frames to the `TacticalVisionNet` inference branch.

```
import cv2
import numpy as np
from PIL import Image

def is_active_gameplay_heuristic(img_path):
    """
    A fast heuristic check: computes the ratio of green pixels in the image.
    If the green ratio exceeds a given threshold, it suggests the screen shows the pitch.
    """
    img = cv2.imread(str(img_path))
    if img is None:
        return False
    
    # Convert to HSV color space
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Define range for pitch green color
    lower_green = np.array([35, 40, 40])
    upper_green = np.array([85, 255, 255])
    
    # Threshold the HSV image to get only green colors
    mask = cv2.inRange(hsv, lower_green, upper_green)
    green_pixels = cv2.countNonZero(mask)
    total_pixels = img.shape[0] * img.shape[1]
    
    green_ratio = green_pixels / total_pixels
    print(f"Green ratio: {green_ratio:.4f}")
    return green_ratio > 0.4  # Gameplay usually has >40% green pitch coverage

print("Gameplay heuristic defined.")
```
