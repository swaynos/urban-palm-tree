# 06_derived_tactical_features.ipynb — Derived Tactical States

Once `TacticalVisionNet` predicts frame-level outputs (possession, pitch zone, and normalized ball coordinates), we compile them over time to compute derived tactical variables. These variables (e.g., ball velocity vectors, team possession transitions, and field zone crossovers) directly inform downstream game strategy controllers.

```
import numpy as np

class TemporalTacticalState:
    """
    Tracks the history of model predictions to calculate velocity vectors
    and detect event transitions (e.g., possession loss, attacking zone entry).
    """
    def __init__(self, buffer_size=10):
        self.buffer_size = buffer_size
        self.possession_history = []
        self.zone_history = []
        self.ball_history = []  # list of (x, y, timestamp)
        
    def update(self, possession, zone, ball_coords, timestamp):
        self.possession_history.append(possession)
        self.zone_history.append(zone)
        self.ball_history.append((ball_coords[0], ball_coords[1], timestamp))
        
        if len(self.possession_history) > self.buffer_size:
            self.possession_history.pop(0)
            self.zone_history.pop(0)
            self.ball_history.pop(0)
            
    def get_ball_velocity(self):
        """Calculate dx/dt and dy/dt of the ball."""
        if len(self.ball_history) < 2:
            return 0.0, 0.0
        
        p1 = self.ball_history[-2]
        p2 = self.ball_history[-1]
        
        dt = p2[2] - p1[2]
        if dt <= 0:
            return 0.0, 0.0
            
        dx = (p2[0] - p1[0]) / dt
        dy = (p2[1] - p1[1]) / dt
        return dx, dy

print("Temporal tracking buffer compiled.")
```
